```python
t = np.linspace(0,10,71)
t
```

```python
f = np.cos(t)
f
```

```python
from pylab import rcParams
dt = pd.DataFrame({'Значения t':t, 'Значения f':f})
dt.set_index('Значения f', inplace = True)
dt
```

```python
rcParams['figure.figsize'] = 20, 20
```

```python
dt.plot(kind="barh")
plt.show()
```

```python
rcParams['figure.figsize'] = 10,10
plt.hist([t,f], ec="green")
plt.axis([0.5,9.5,-2.5,2.5])
plt.show()
```
